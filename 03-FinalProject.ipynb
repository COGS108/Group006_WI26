## Abstract

This study investigates the relationship between state-level socioeconomic conditions and academic performance among U.S. eighth-grade students during the 2018–2019 school year. Motivated by longstanding evidence that socioeconomic status influences educational outcomes, we specifically examine whether higher median household income is associated with higher average student test scores. Focusing on eighth grade—a critical transition point before high school—allows us to better understand how economic disparities may shape early academic trajectories.

To address this question, we combined academic performance data from the Stanford Education Data Archive (SEDA) with socioeconomic indicators at the school district level. After cleaning and merging the datasets, we conducted exploratory data analysis, including correlation matrices and distribution visualizations, to identify patterns between variables. We then applied linear regression models to quantify the relationships between test scores (Mean GCS scores) and key predictors such as median household income, poverty rate, and the proportion of students receiving free or reduced lunch.

Our findings indicate a consistent and statistically significant positive relationship between median household income and academic performance, while measures of economic disadvantage—particularly the percentage of students receiving free or reduced lunch—show strong negative associations with test scores. These trends were observed across both mathematics and reading/language arts subjects. The regression models explained a substantial portion of the variance in test scores (R² ≈ 0.49), suggesting that socioeconomic factors play a major, though not exclusive, role in shaping student achievement.

Overall, the results support the hypothesis that higher-income environments are linked to better academic outcomes and highlight the persistent impact of socioeconomic inequality on education. While the analysis is limited by its cross-sectional design and aggregated data, it underscores the importance of addressing economic disparities to promote more equitable educational opportunities.

Disparities in academic performance amongst students have long been attributed to differences in socioeconomic status among American students. Since the transition from middle school to high school marks the first step to adulthood, we want to look at the the academic performance for 8th graders and compare it to the socioeconmic status of their household. It is most often believed that students that perfrom well during 8th grade, tend to continue to perform well in highschool and further. Seeing as 8th grade has public testing and more accessible data, we will choose to analyze the relationship between socioeconomic status and academic performance within 8th graders.

A similar yet older study by The American Physiological Association analyzed the relationship between socioeconomic status and education. This dataset is the SEDA (Stanford Education Data Archive) Administrative District Test Score dataset (version 5.0), organized in a long format where each row corresponds to a particular school administrative district in a given year (from 2009 to 2019), grade level (from 3 to 8). According to our research needs, we can filter out the necessary information. The dataset contains 1048575 observations and 56 variables. Districts are identified using a unique district code (sedaadmin), with geographic identifiers including state abbreviation (stateabb) and FIPS state code.

The major group of the dataset is students’ academic performance measured through GCS (Grade Cohort Standardized) scores. The variable gcs_mn_all represents the mean GCS score for students. GCS scores allow comparisons across grade levels, with higher values indicating better academic performance. The variable tot_asmt_all indicates the total number of students assessed, providing the sample size for the analysis.

Another feature of this dataset is that it groups students into groups relevant to economically disadvantaged, which is relevant to our research question on relationships between income and academic performances. The variable gcs_mn_ecd represents the mean score for economically disadvantaged students, and gcs_mn_nec represents the mean score for non-economically disadvantaged students. These two variables allow us to determine the relationship between students’ academic performance in different economic classes within the same and different states.

There are several limitations of this dataset. First, the dataset does not contain direct household income data. It only provides a binary classification of students as economically disadvantaged or not. This cannot completely replace the income data. Second, there are missing values in gcs_mn_ecd and gcs_mn_nec, which our research needs to focus on.

This dataset is the SEDA (Stanford Education Data Archive) Administrative District Covariates dataset (version 5.0), organized in a long format where each row corresponds to a particular school administrative district in a given year and grade level (grades 3 through 8, for school years 2008–09 through 2018–19). Districts are identified using a unique district code (sedaadmin) and a district name (sedaadminname). The dataset also includes geographic identifiers such as FIPS state code (fips) and district location descriptors. Overall, the dataset is designed to provide background socioeconomic and demographic context for analyzing educational outcomes across U.S. school districts.

A major group of variables describes district socioeconomic conditions, largely derived from the American Community Survey (ACS). For example, the dataset contains poverty rate measures (proportion of households below the poverty line), single-parent household rates, and adult educational attainment rates. These variables are expressed as proportions (0 to 1), and they capture the social environment students are likely growing up in. For example, a “poverty rate” of 0.25 would mean that about 25% of households in that district fall below the poverty threshold. These socioeconomic measures matter because districts with higher poverty and lower adult education levels often face fewer educational resources, which can strongly influence student achievement patterns.

The dataset also contains urbanicity information (urbanicity), which describes whether a district is urban, suburban, town-based, or rural (based on student enrollment). This variable is categorical rather than numeric, and it provides important context because school funding, teacher availability, and student access to academic resources often differ dramatically across rural vs. urban districts. In educational research, urbanicity is often a key explanatory factor because it correlates with infrastructure access, local tax bases, and student population density.

A major concern with this dataset is that it is not individual-level student data. Instead, it is aggregated at the district level, meaning any conclusions drawn from it risk ecological fallacy (assuming that relationships observed at the district level necessarily apply to individual students). Another limitation is that many variables come from ACS estimates, which are subject to sampling error, especially for smaller districts. Some values may be produced using modeling or smoothing methods (such as “EB estimates,” empirical Bayes), meaning they are not purely raw observed measurements. This can be helpful for stability, but it also means the dataset contains estimates rather than exact truth.

3. Three columns (`Percent_Free_Lunch`, `Percent_Reduced_Lunch`, `Percent_FRL_Total`)

Since the six socioeconomic columns share the same set of missing rows, they should lack the same socioeconomic datapoints. Omitting 1140 individuals from our observed population due to empty datapoints may cause bias if they are coming from some specific school districts that do not collect parental information.

Filling the missing values with median:

```python
df_clean_processed = df_clean.copy()

cols_with_missing = ['Median_Income_Log', 'Percent_BA_Degree', 'Poverty_Rate', \
                     'Unemployment_Rate', 'Percent_SNAP', 'Percent_Single_Moms', \
                     'Percent_Econ_Disadv', 'Percent_Free_Lunch', \
                     'Percent_Reduced_Lunch', 'Percent_FRL_Total']

for col in cols_with_missing:
    median_val = df_clean_processed[col].median()
    df_clean_processed[col] = df_clean_processed[col].fillna(median_val)

# let's make sure that we still have our entire dataset
clean_shape_pro = df_clean_processed.shape
print(f'This dataframe has {clean_shape_pro[0]:,} rows and {clean_shape_pro[1]} columns.')
display(df_clean_processed.head())
```

## Discusison and Conclusion

The objective of this project is to investigate the relationship between the socio-economic conditions of various states in the United States (particularly the median household income) and the academic performance of eighth-grade students in the 2018-2019 academic year. We utilized the SEDA dataset and its socio-economic covariates to analyze the connection between test scores (Mean GCS scores) and various economic indicators, including the median household income, poverty rate, and the proportion of students receiving free/reduced-price lunches. Through exploratory data analysis and regression modeling, our results generally support the initial hypothesis: regions with higher income levels typically correspond to higher student academic performance.

Exploratory data analysis revealed a clear pattern between socio-economic variables and test scores. The correlation matrix showed that the median household income (in logarithmic form) and the proportion of adults with a bachelor's degree were positively correlated with students' scores, while variables related to economic hardship (such as poverty rate, proportion of economically disadvantaged students, and proportion of free/reduced-price lunch) were negatively correlated with scores. These patterns were consistent in both mathematics and English (reading/language arts) subjects, indicating that the influence of the socio-economic environment on students' performance is universal and not limited to a specific subject.

Further linear regression analysis strengthened the above findings. In the mathematics and English models, the median household income had a significant positive impact on academic performance, while the proportion of free/reduced-price lunches showed a significant negative impact. The explanatory power of the model was approximately half (R² ≈ 0.49), indicating that socio-economic factors are important predictors of academic performance, but not the only factors. It is notable that the poverty rate did not show statistical significance in the regression model, which may imply that other variables (such as access to educational resources, school quality, or parental education level) partially substitute for or absorb its influence.

Overall, these results expand the research conclusions in the background literature regarding the relationship between socioeconomic status and educational outcomes. Our analysis supports the viewpoints of previous studies: wealthier regions typically provide students with more resources, such as better educational materials, extracurricular tutoring opportunities, and more stable family environments. In contrast, in economically disadvantaged areas, students may face structural barriers, such as insufficient school funds, food insecurity, and a lack of academic support systems, which can have a negative impact on academic performance. By focusing on the critical transitional stage of eighth grade (before entering high school), this study further emphasizes how economic disparities have a significant impact on students' educational trajectories in the early stages.

Nevertheless, this study still has several limitations. Firstly, the data is only cross-sectional data for the 2018-2019 academic year, so we cannot establish causal relationships or observe the changes of these patterns over time. If longitudinal data could be used to track the same regions or student groups, it would be more helpful in revealing the long-term impact of socio-economic factors on academic performance. Additionally, most of the socio-economic indicators in this study are based on aggregated data at the school district level, which may mask the differences at the community or family level. Finally, although the data set contains various economic variables, it fails to directly reflect other key factors such as education policies, teacher quality, or family education support, which may also have significant impacts on student performance.

Future research can be further expanded in multiple directions. For instance, longitudinal data over multiple years can be introduced to analyze whether changes in economic conditions lead to corresponding changes in academic performance; studies can also be conducted on regional differences or educational policy interventions to explore which educational systems are more effective in alleviating social and economic inequalities. In summary, a deep understanding of the interaction between economic conditions and the education system is of great significance for formulating policies to promote educational equity.
