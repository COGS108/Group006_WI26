## Research Question

How does the state-level median household income affect the academic performance (numerical average grades) for 8th grade students across the United States during 2018 - 2019 (before COVID-19 pandemic)?

## Background and Prior Work

Disparities in academic performance amongst students have long been attributed to differences in socioeconomic status among American students. Since the transition from middle school to high school marks the first step to adulthood, we want to look at the the academic performance for 8th graders and compare it to the socioeconmic status of their household. It is most often believed that students that perfrom well during 8th grade, tend to continue to perform well in highschool and further. Seeing as 8th grade has public testing and more accessible data, we will choose to analyze the relationship between socioeconomic status and academic performance within 8th graders

A similar yet older study by The American Physiological Association analyzed the relationship between socioeconomic status and education. This study deemed, “Individuals within the top family income quartile are 8 times more likely to obtain a bachelor’s degree by age 24 as compared to individuals from the lowest family income quartile”<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2).

According to the National Assessment of Educational Progress (NAEP), 9 years old and 13 years old students in the United States have an increase in average scores for NAEP long-term trend (LTT) reading and mathematics assessments from 2018-2019 <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Meanwhile, with the rise of academic performance, the data from the United States Census Bureau has informed us that there is a 6.8 percent increase (from 64,323 to 68,703 dollars) in household income in the United States <a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7).

Maria N. Wodarz, Natalia L. Komarova and Timmy Ma had done research for the relationship of English Language Arts (ELA) and Mathematics performance scores in 3rd and 8th grades and the median household income from during the 2018-2019 school year. The research found out, in half of the states, there is a linear correlation between academic performance and household income, and the other half characterized by a saturating function. The results reveal that less affluent states exhibit a steeper increase in performance with income compared to wealthier states <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). 

With the inquiry of the relationship between median household income and academic performance in 3rd and 8th grades, it is important to focus specifically on 8th grade students, as this stage represents a major academic and developmental transition. During 8th grade, students are preparing to enter high school, and their academic achievement can influence placement into advanced courses, access to academic resources, and long-term educational opportunities. By analyzing this relationship, this research can help identify whether economic differences between states are strongly associated with measurable gaps in academic performance.

1. <a name="cite_note-1"></a> [^](#cite_ref-1) Wodarz, Maria N., et al. “The Functional Form of the Association between K-12 Student Performance and Household Income in U.S. School Districts.” PLOS ONE, Public Library of Science, 10 Sept. 2025, journals.plos.org/plosone/article?id=10.1371%2Fjournal.pone.0329296.   
2. <a name="cite_note-2"></a> [^](#cite_ref-2) “Education and Socioeconomic Status.” American Psychological Association, American Psychological Association, 2017, www.apa.org/pi/ses/resources/publications/education. 
3. <a name="cite_note-3"></a> [^](#cite_ref-3) “NAEP Long-Term Trend Assessment Results:Reading and Mathematics.” The Nation’s Report Card, 2023, www.nationsreportcard.gov/ltt/?age=9. 
3. <a name="cite_note-4"></a> [^](#cite_ref-4) Semega, Jessica. “Income and Poverty in the United States: 2019.” Census.Gov, 9 June 2022, www.census.gov/library/publications/2020/demo/p60-270.html.

## Hypothesis


There is a positive correlation between state-level median household income and academic performance (average grades) among U.S. 8th grade students during 2018-2019, such that states with higher median household income tend to have higher average student grades. We believe that higher household income means access to better academic resources and financial improvement shapes students self-confidence, therefore, with the increase of household income, students would have a better academic performance. We also believe that students that test better in 8th grade tend to have better academic success in high school.

### Data overview

#### Dataset 1: SEDA (Stanford Education Data Archive) Administrative District Test Score Dataset (version 5.0)
- Dataset Name: seda_admindist_long_gcs.csv
- Link to Dataset: https://stacks.stanford.edu/file/druid:cs829jn7849/seda_admindist_long_gcs_5.0_updated_20240319.csv
- Number of Observations: 1,048,575
- Number of Variables: 56
- Description of Variables:
    - stateabb - state abbreviation, which can be used in state-level comparison year: academic year, including years between 2009 and 2018 (inclusive), can be used to filter out year 2018 to 2019.
    - grade - grade level, from grade 3 to grade 8. We are including grade 8 instead of senior high school students.
    - subject - the subject that the students are studying. It can be used to group students and compare their academic performance.
    - tot_asmt_all - the total number of students assessed. It is the sample size of different districts.
    - gcs_mn_all: the mean GCS score for students. This is the dependent variable that represents academic performance.
    - gcs_mn_ecd: the mean score for economically disadvantaged students.
    - gcs_mn_nec: the mean score for non-economically disadvantaged students.
        - The above variables will help filter between 2018-2019, exclusively 8th grade, and compare scores based on income levels.
- Description of Shortcomings:
    - The income data is not included. This dataset only includes students’ academic performance, and divides the students into economically disadvantaged students and non_economically disadvantaged students. There are missing values in the variables related to our project. For example, gcs_mn_ecd and gcs_mn_nec.

#### Dataset 2: SEDA (Stanford Education Data Archive) Administrative Disctrict Covariates Dataset (version 5.0)
- Dataset Name: seda_cov_admindist_long_5.0*
- Link to Dataset: https://stacks.stanford.edu/file/druid:cs829jn7849/seda_cov_admindist_long_5.0_updated_20240319.csv
- Number of Observations: 1,000,000
- Number of Variables: 73
- Description of Variables:
    - lninc50all: log of median income, eb estimate, all families, time-varying. This represents the log-transformed median household income (median income across families). This is the closest match to state-level median household income in the dataset.
    - povertyall: poverty rate, eb estimate, all families, time-varying. This represent the overall poverty rate. This is another variable which we could consider to determine the relationship between academic performance and household income.
    - year and grade: These two variables ensure you are analyzing 8th graders in the 2018–2019 school year.
    - urbanicity: urbanicity locale (based on student enrollment).This could be considered as one of the factors affecting the household income.
    - fips and stateabb: These allow you to group by state to create state-level averages.
    - There are also income and poverty data for different racial which could also be considered while analyzing.
- Description of Shortcomings:
    - The data is aggregated at the district level, not individual-level student data.
    - Many variables came from ACS,  some values may be produced using modeling or smoothing methods (such as “EB estimates,” empirical Bayes).

How do we plan on combing these?
- We plan on combining these by using school district id's (sedaadmin). This will allow us to match up the scores of the 8th graders and the socioeconmic status that their school districts fall under.
