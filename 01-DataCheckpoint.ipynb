## Research Question

How does the state-level median household income affect the academic performance (numerical average grades) for senior high school students across the United States during 2018 - 2019 (before COVID-19 pandemic)?

## Background and Prior Work

Disparities in academic performance amongst students have long been attributed to differences in socioeconomic status among American students. Since senior year in high school marks the turning point in becoming an adult, we want to analyze the true disparity between the academic performance of high school seniors based on socioeconomic status to see if a relationship exists.

A similar yet older study by The American Physiological Association  analyzed the relationship between socioeconomic status and education. This study deemed, “Individuals within the top family income quartile are 8 times more likely to obtain a bachelor’s degree by age 24 as compared to individuals from the lowest family income quartile”<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2).

According to the National Assessment of Educational Progress (NAEP), 9 years old and 13 years old students in the United States have an increase in average scores for NAEP long-term trend (LTT) reading and mathematics assessments from 2018-2019 <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Meanwhile, with the rise of academic performance, the data from the United States Census Bureau has informed us that there is a 6.8 percent increase (from $64,324 to $68,703) in household income in the United States <a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7).

Maria N. Wodarz, Natalia L. Komarova and Timmy Ma had done research for the relationship of English Language Arts (ELA) and Mathematics performance scores in 3rd and 8th grades and the median household income from during the 2018-2019 school year. The research found out, in half of the states, there is a linear correlation between academic performance and household income, and the other half characterized by a saturating function. The results reveal that less affluent states exhibit a steeper increase in performance with income compared to wealthier states <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). 

With the inquiry of the relationship for students in grade three and grade eight, it is important to examine whether this association persists for students in 12th grade. This stage represents a critical transition period, as many students are preparing for graduation and applying to undergraduate institutions. Since academic performance during the final year of high school can strongly influence college admissions opportunities, investigating the impact of household income on 12th grade achievement may provide deeper insight into how socioeconomic disparities shape educational outcomes at the point of entry into higher education.

1. <a name="cite_note-1"></a> [^](#cite_ref-1) Wodarz, Maria N., et al. “The Functional Form of the Association between K-12 Student Performance and Household Income in U.S. School Districts.” PLOS ONE, Public Library of Science, 10 Sept. 2025, www.journals.plos.org/plosone/article?id=10.1371%2Fjournal.pone.0329296.   
2. <a name="cite_note-2"></a> [^](#cite_ref-2) “Education and Socioeconomic Status.” American Psychological Association, American Psychological Association, 2017, www.apa.org/pi/ses/resources/publications/education. 
3. <a name="cite_note-3"></a> [^](#cite_ref-3) “NAEP Long-Term Trend Assessment Results:Reading and Mathematics.” The Nation’s Report Card, 2023, www.nationsreportcard.gov/ltt/?age=9. 
3. <a name="cite_note-4"></a> [^](#cite_ref-4) Semega, Jessica. “Income and Poverty in the United States: 2019.” Census.Gov, 9 June 2022, www.census.gov/library/publications/2020/demo/p60-270.html.

## Hypothesis


There is a positive correlation between state-level median household income and academic performance (average grades) among U.S. senior high school students during 2018-2019, such that states with higher median household income tend to have higher average student grades. We believe that higher household income means access to better academic resources and financial improvement shapes students self-confidence, therefore, with the increase of household income, students would have a better academic performance.
