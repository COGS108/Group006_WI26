# Research Question

How does the state-level median household income affect the academic performance (numerical average grades) for 8th grade students across the United States during 2018 - 2019 (before COVID-19 pandemic)?

## Background and Prior Work

Disparities in academic performance amongst students have long been attributed to differences in socioeconomic status among American students. Since the transition from middle school to high school marks the first step to adulthood, we want to look at the the academic performance for 8th graders and compare it to the socioeconmic status of their household. It is most often believed that students that perfrom well during 8th grade, tend to continue to perform well in highschool and further. Seeing as 8th grade has public testing and more accessible data, we will choose to analyze the relationship between socioeconomic status and academic performance within 8th graders.

A similar yet older study by The American Physiological Association analyzed the relationship between socioeconomic status and education. This study deemed, “Individuals within the top family income quartile are 8 times more likely to obtain a bachelor’s degree by age 24 as compared to individuals from the lowest family income quartile”2.

According to the National Assessment of Educational Progress (NAEP), 9 years old and 13 years old students in the United States have an increase in average scores for NAEP long-term trend (LTT) reading and mathematics assessments from 2018-2019 3. Meanwhile, with the rise of academic performance, the data from the United States Census Bureau has informed us that there is a 6.8 percent increase (from 64,323 to 68,703 dollars) in household income in the United States 7.

Maria N. Wodarz, Natalia L. Komarova and Timmy Ma had done research for the relationship of English Language Arts (ELA) and Mathematics performance scores in 3rd and 8th grades and the median household income from during the 2018-2019 school year. The research found out, in half of the states, there is a linear correlation between academic performance and household income, and the other half characterized by a saturating function. The results reveal that less affluent states exhibit a steeper increase in performance with income compared to wealthier states 1.

With the inquiry of the relationship between median household income and academic performance in 3rd and 8th grades, it is important to focus specifically on 8th grade students, as this stage represents a major academic and developmental transition. During 8th grade, students are preparing to enter high school, and their academic achievement can influence placement into advanced courses, access to academic resources, and long-term educational opportunities. By analyzing this relationship, this research can help identify whether economic differences between states are strongly associated with measurable gaps in academic performance.

^ Wodarz, Maria N., et al. “The Functional Form of the Association between K-12 Student Performance and Household Income in U.S. School Districts.” PLOS ONE, Public Library of Science, 10 Sept. 2025, journals.plos.org/plosone/article?id=10.1371%2Fjournal.pone.0329296.
^ “Education and Socioeconomic Status.” American Psychological Association, American Psychological Association, 2017, www.apa.org/pi/ses/resources/publications/education.

There is a positive correlation between state-level median household income and academic performance (average grades) among U.S. 8th grade students during 2018-2019, such that states with higher median household income tend to have higher average student grades. #### Missing Data
Now to ensure that our combined dataset is fully wrangled, cleaned, and tidied. Let us clear up any of the remaining NaN values that we have so the we can work with our dataset.

Let's take a look at how much missing data we have:

```python
missing_values = df_clean.isnull().sum()
missing_percent = (missing_values / len(df_clean)) * 100

missing_df = pd.DataFrame({'Missing Count': missing_values})

missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)
display(missing_df)
```

From the dataframe above, we can see that:

1. Six socioeconomic columns (`Median_Income_Log`, `Percent_BA_Degree`, `Poverty_Rate`, `Unemployment_Rate`, `Percent_SNAP`, `Percent_Single_Moms`) have the same 1140 missing counts.
2. 'Percent_Econ_Disadv has 156 missing values.
3. Three columns (`Percent_Free_Lunch`, `Percent_Reduced_Lunch`, `Percent_FRL_Total`)

Since the six socioeconomic columns share the same set of missing rows, they should lack the same socioeconomic datapoints. Omitting 1140 individuals from our observed population due to empty datapoints may cause bias if they are coming from some specific school districts that do not collect parental information.

Therefore, We choose to fill these missing values with the median of each column, as this value is less affected by extreme outliers and more representative of the 'average American household'.

Filling the missing values with median:

```python
df_clean_processed = df_clean.copy()

cols_with_missing = ['Median_Income_Log', 'Percent_BA_Degree', 'Poverty_Rate', \
                     'Unemployment_Rate', 'Percent_SNAP', 'Percent_Single_Moms', \
                     'Percent_Econ_Disadv', 'Percent_Free_Lunch', \
                     'Percent_Reduced_Lunch', 'Percent_FRL_Total']

for col in cols_with_missing:
    median_val = df_clean_processed[col].median()
    df_clean_processed[col] = df_clean_processed[col].fillna(median_val)

df_clean_processed.head()
```

Verifying that there is no missing values in the dataframe:

```python
print(f"Remaining missing values: {df_clean_processed.isnull().sum().sum()}")
```
